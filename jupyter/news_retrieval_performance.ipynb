```python
import sys
sys.path.append('../py/')
from news_sampler import news_sampling, calculate_idf

import glob
corpus_fnames = glob.glob('/mnt/sdc1/papers/cohesion/models/2016-10-20_article_all_normed_*')
index_fname = '/mnt/sdc1/papers/data/2016-10-20_article_all.index'
tokenizer_names = [fname.split('/')[-1].split('.')[0].split('_normed_')[-1] for fname in corpus_fnames]
```

```python
def get_keywords(doc, n_keywords=5):
    terms = doc.split()
    terms = [term.split('//') for term in terms]
    terms = [(term[0], int(term[1])) for term in terms if len(term) == 2]
    terms = [(term[0].split('/')[0], term[1]) for term in terms if (not '/' in term[0]) or ('/N' in term[0])]
    return sorted(terms, key=lambda x:x[1], reverse=True)[:n_keywords]
    
sampler = news_sampling(corpus_fnames, index_fname, 50)
for i, docs in enumerate(sampler):
    index = docs[-1]
    for t, doc in enumerate(docs[:-1]):
        print('%20s'%tokenizer_names[t], get_keywords(doc))
    print(index)
    print()
```
